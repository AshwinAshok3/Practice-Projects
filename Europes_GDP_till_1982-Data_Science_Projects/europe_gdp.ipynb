```python
# i personally think that this data is actually sorted though we will still sort this by country 
# using the groupby function in pandas library

europe_gdp = europe_pd.groupby('country').sum().reset_index()
```

```python
# for this we will create a function for choosing random light colours 
# which are unique in nature and we will use colors library from matplotlib
```

```python
import numpy as np
import colorsys as clrys
```

```python
# since we have about 30 countries to select from naming those 30 countries would be time consuming 
# because imagine doing so for all the 200 countries around the world wouldn't it be tiresome
# therefore i think , to have a function that would pick light colours which are unique in nature would do so .
```

```python
def random_unique_colors(n):
    colors = set()    # so that our colours will have uniqueness
    
    # the use of while loop is done because since we are using a set 
    # if we encounter same hex value of colours then it would automatically get's cancelled so it won't duplicate 
    while len(colors) < n:
        
        #Generate a random color
        # also ensuring that these ranges are only in between (0,1)
        h = np.random.rand()
        l = 0.7 + 0.3 * np.random.rand()
        s = 0.5 + 0.5 * np.random.rand()
        
        r,g,b = clrys.hls_to_rgb(h,l,s)
        
        light_color = (r,g,b)
        
        colors.add(light_color)
        
    return list(colors)
```

```python
colours = random_unique_colors(len(europe_gdp))
```

```python
# now we have the hex values of the light colours , 
# also doesn't matter if they are not strings it would still take the values.
```

```python
colours
```

```python
# defining the explode

# the explode will contain such values as (making value as 0.1 else 0) when the % is <= 3.5 
#where 3.5 is the threshold that we consider
# fill the explode as list, as this list contains the size on the dataframe column-row.
# Even so , we still have to calculate the total number in percentage of 100 because the europe_gdp['Total'] == price values

sum_of_total = europe_gdp['Total'].sum()
# we found the total sum of the column because to find the percentage of that column in 100% 
```

```python
sum_of_total
```

```python
limit = 3.5
```

```python
# here we find the percentages out of 100 and make it a new column , series in the dataframe europe_gdp 
# if you wanna change series into something else please feel free to do so .
```

```python
europe_gdp['values_in_100s'] = (europe_gdp['Total']/sum_of_total)*100
```

```python
europe_gdp['values_in_100s']
```

```python
europe_gdp.columns
```

```python
# using list comprehension we can satisfy our condition on naming the explodes as 0.1 
# where the percent is actually less than the threshold.

explode = [0.1 if _ < limit else 0 for _ in europe_gdp['values_in_100s']]
```

```python
explode
```

```python
import matplotlib.patheffects as pe
```

```python

fig , ax = plt.subplots(figsize=(17,15))

wedges ,texts, autotexts = ax.pie(
    europe_gdp['Total'],
    autopct = '%1.1f%%' ,
    startangle = 180,
    labels = europe_gdp['country'] ,
    pctdistance = 1.05,
    colors = colours,
    explode = explode,
    shadow = True
)

# Adding 3D effect to each slice
for wedge in wedges:
    wedge.set_edgecolor('black')
    wedge.set_linewidth(1.5)
    wedge.set_path_effects([pe.SimpleLineShadow(), pe.Normal()])


plt.title('Europes GDP from 1952 to 2007')
plt.axis('equal')
plt.legend(labels = europe_gdp['country'].values,
          loc = (1,0) )
plt.show()
```
