## Importing Libraries

````python
# import libraries
import numpy as np   
import pandas as pd   
import matplotlib.pyplot as plt  
import seaborn as sns 
%matplotlib inline 
````

## Reading Data Set

````python
# Read the data set using read_csv function

df=pd.read_csv("Covid_19_cases.csv")
````

### Get glimpse of data

````python
df.head()
````

````python
# consider the columns 'cases', 'deaths', 'countriesAndTerritories' and prepare a new dataframe
# Write your code here

df_new = df[['cases','deaths','countriesAndTerritories']]
````

### Q1. Use groupby on <code>countriesAndTerritories</code> to find total <code>cases</code> and <code>deaths</code>

````python
# Write your code here
new_df_CnT = df_new.groupby('countriesAndTerritories')[['cases','deaths']].sum().reset_index()
````

````python
new_df_CnT
````

### Q2. Find top five countries in terms of <code>cases</code>, store them in a new dataframe and Visualize them

````python
# Write your code here

new_countries_desc = new_df_CnT.sort_values(by='cases' , ascending = False )
````

````python
N_C_desc = new_countries_desc[0:5]
````

````python
N_C_desc
````

````python
# in terms of cases

sns.barplot(x='countriesAndTerritories',
            y='cases',
            data = N_C_desc)

plt.xlabel('Countries')
plt.ylabel('Cases')
plt.title('Countries By Covid Cases')
````

### Q3. Find top five countries in terms of <code>deaths</code>, store them in a new dataframe and Visualize them

````python
# Write your code here

new_countries_dt_desc = new_df_CnT.sort_values(by='deaths' , ascending = False )
````

````python
N_D_desc = new_countries_dt_desc[0:5]
````

````python
N_D_desc
````

````python
# in terms of cases

sns.barplot(x='countriesAndTerritories',
            y='deaths',
            data = N_D_desc)

plt.xlabel('Countries')
plt.ylabel('No Of Deaths')
plt.title('Top 5 Countries By Covid Deaths')
````

### Q4. Use groupby on <code>countriesAndTerritories</code> to find mean value of <code>cases</code> and <code>deaths</code>

````python
# Write your code here

mean_df_CnT = new_df_CnT.groupby('countriesAndTerritories')[['cases','deaths']].mean().reset_index()

````

````python
mean_df_CnT
````

### Q5. Find top five countries with highest number of average <code>cases</code>, store them in a new dataframe and Visualize them

````python
# Write your code here

mean_df_C = new_df_CnT.groupby('countriesAndTerritories')[['cases']].mean().reset_index()
````

````python
new_mean_country = mean_df_C.sort_values(by='cases', ascending=False)
````

````python
top_five_by_mean_cases = new_mean_country[0:5]
````

````python
top_five_by_mean_cases
````

````python
# colours = ['red','green','yellow','blue','purple']
````

````python

sns.barplot(
    x='countriesAndTerritories' ,
    y='cases',
    data = top_five_by_mean_cases)


plt.xlabel('Countries')
plt.ylabel('Average Cases')
plt.xticks(rotation=35 , fontsize=9)
plt.title('Average Covid Cases of Top 5 countries')
````

### Q6. Find top five countries with highest number of average <code>deaths</code>, store them in a new dataframe and Visualize them
