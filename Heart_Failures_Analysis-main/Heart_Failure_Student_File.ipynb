### Q1. Create new dataframe <code>df_old</code> with <code>age>=60</code> and replace all the values in <code>age</code> column of <code>df_old</code> with <code>old age</code>

````python
###### Write your code here


# Here in the new variable 'df_old' we will store our data of people according their age if above or equal to '60'
````

````python

df_old = heart_failures_df[ heart_failures_df['age'] >=60 ]
````

````python
df_old
````

````python
# we have renamed the name of the column 'age' to 'old_age' 
# because these dataframe consists of people who are above the age of 60

df_old.rename(columns={'age':'old_age'} , inplace = True)
````

### Q2. Create new dataframe <code>df_middle_age</code> with <code>age</code>< 60 and replace all the values in <code>age</code> column of <code>df_middle_age</code> with <code>middle age</code>

````python
# Write your code here


# Here we will repeat the same process to the same question 
# however we will adjust the condition inside the dataframe to be '< 60'
````

````python
middle_age = heart_failures_df[heart_failures_df['age'] < 60]
````

````python

middle_age.rename(columns={'age':'middle_age'}, inplace = True)
````

````python
middle_age
````

### Q3. Concatenate <code>df_old</code> and <code>df_middle_age</code> and form a new dataframe as <code>df_new</code>

````python
# Hint: use pd.concat()

# to add both the data frame into a new dataframe ,
# however this concat() doesn't overwrite over other dataframe which is actually good

df_new = pd.concat([df_old,middle_age])
````

````python
df_new.columns
````

### Q4. Visualize <code>diabetes, platelets, DEATH_EVENT</code> for <code>middle age</code> and <code>old age</code> people

````python
# Hint: sns.barplot(x='',y='',data=)

# To barplot 'diabetes, platelets, DEATH_EVENT' these columns for middle age and old age .
# I will be creating two dataframe for middle age->  'D_P_DE_middle_age_df' and 
# old age -> 'D_P_DE_old_age_df' so to avoid any confusions.
````

````python

D_P_DE_old_age_df = df_new.groupby('old_age')[['diabetes','platelets','DEATH_EVENT']].mean().reset_index()

D_P_DE_middle_age_df = df_new.groupby('middle_age')[['diabetes','platelets','DEATH_EVENT']].mean().reset_index()
````

````python
D_P_DE_old_age_df.columns
````

````python
D_P_DE_middle_age_df.columns
````

````python
# Diabetic patients aged above 60 and 60


sns.barplot(
    width = 0.85,
    x = 'old_age',
    y = 'diabetes',
    data = D_P_DE_old_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Diabetes')
plt.title('Diabetic Individuals (Old Age)')

````

````python
# Diabetic patients aged below 60


sns.barplot(
    width = 0.85,
    x = 'middle_age',
    y = 'diabetes',
    data = D_P_DE_middle_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Diabetes')
plt.title('Diabetic Individuals (Middle Age)')
````

````python
# platlets counts of people aged above 60 and 60


sns.barplot(
    width = 0.85,
    x = 'old_age',
    y = 'platelets',
    data = D_P_DE_old_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Platelets Counts')
plt.title('Platelets Count (Old Age)')
plt.show()
````

````python
# platlets counts of people aged below 60 


sns.barplot(
    width = 0.85,
    x = 'middle_age',
    y = 'platelets',
    data = D_P_DE_middle_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Platelets Counts ')
plt.title('Platelets Count (Middle Age)')
plt.show()
````

````python
# No.of deaths of people of age above 60 and 60

sns.barplot(
    width = 0.85,
    x = 'old_age',
    y = 'DEATH_EVENT',
    data = D_P_DE_old_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Deaths')
plt.title(' Deaths (Old Age)')
plt.show()
````

````python
# No.of deaths of people of age velow 60 


sns.barplot(
    width = 0.85,
    x = 'middle_age',
    y = 'DEATH_EVENT',
    data = D_P_DE_middle_age_df)

plt.xticks(rotation=90 , fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Deaths')
plt.title(' Deaths (Middle Age)')
plt.show()

````

### Q6. Visualize average <code>platelets</code> for <code>middle age</code> and <code>old age</code> people

````python
# Write your code here

# here we will check the platelet counts for both middle age and old age
# by using two new dataframes for middle and old each.
````

````python

Platelets_old_age_df = df_new.groupby('old_age')[['platelets']].mean().reset_index()

Platelets_middle_age_df = df_new.groupby('middle_age')[['platelets']].mean().reset_index()
````

````python

sns.barplot(
    x='middle_age',
    y='platelets',
    data = Platelets_middle_age_df
)

plt.xticks(rotation = 90, fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Platelets')
plt.title('Average Platelet Count By Age of middle aged')


````

````python

sns.barplot(
    x='old_age',
    y='platelets',
    data = Platelets_old_age_df
)

plt.xticks(rotation = 90, fontsize = 8)
plt.xlabel('AGE')
plt.ylabel('Platelets')
plt.title('Average Platelet Count By Age of Old Aged')


````
